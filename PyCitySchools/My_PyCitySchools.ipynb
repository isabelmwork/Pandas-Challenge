# PyCity Schools Analysis

Overall, the high-performing schools tend to be charter schools with lower per-student budgets and school sizes. On the other hand, low-performing schools tend to be district schools with higher per-student budgets and school sizes. 

Average reading score stays consistent regardless of school type and other factors. Similarly, average math and reading scores are consistent throughout grade levels. 

Charter schools have about a 90% overall pass rate, while district schools have about a 53% pass rate. 

```python
import pandas as pd
from pathlib import Path

school_path = Path("Resources/schools_complete.csv")
student_path = Path("Resources/students_complete.csv")
```

```python
#Read and output school_data

school_data = pd.read_csv(school_path)
school_data.head()
```

```python
#Read and output student_data

student_data = pd.read_csv(student_path)
student_data.head()
```

```python
#Merge school_data and student_data by "school_name"
#Output merged data

complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_data.head()
```

## District Summary

```python
#Total number of unique schools

total_unique_schools = len(student_data["school_name"].unique())
total_unique_schools
```

```python
#Total students

total_students = len(student_data["Student ID"].unique())
total_students
```

```python
#Total budget

total_budget = sum(school_data["budget"])
total_budget
```

```python
#Average math score

avg_math_score = student_data["math_score"].mean()
avg_math_score
```

```python
#Average reading score

avg_reading_score = student_data["reading_score"].mean()
avg_reading_score
```

```python
#% passing math
#used .loc to extract data of students passing math

students_passing_math = len(student_data.loc[student_data["math_score"] >= 70])
perc_passing_math = (students_passing_math / total_students) * 100
perc_passing_math
```

```python
#% passing reading
#used .loc to extract data of students passing reading

students_passing_reading = len(student_data.loc[student_data["reading_score"] >= 70])
perc_passing_reading = (students_passing_reading / total_students) * 100
perc_passing_reading
```

```python
#% passing overall
#used .loc to extract data of students passing overall
#used & operator to extract data that meets both conditions

students_passing = len(student_data.loc[(student_data["math_score"] >= 70) & 
                                         (student_data["reading_score"] >= 70)])
perc_passing = (students_passing / total_students) * 100
perc_passing
```

```python
#created dataframe with column names
#used .loc to input data

district_snapshot = pd.DataFrame(columns = ["Total_Schools", "Total_Students", "Total_Budget", "Average_Math_Score", "Average_Reading_Score", "%_Passing_Math", "%_Passing_Reading", "%_Passing_Overall"])
district_snapshot.loc[0] = [total_unique_schools, total_students, total_budget, avg_math_score, avg_reading_score, perc_passing_math, perc_passing_reading, perc_passing]

district_snapshot
```

```python
#formatted the data
district_snapshot["Total_Budget"] = district_snapshot["Total_Budget"].map("${:,.2f}".format)
district_snapshot["Total_Students"] = district_snapshot["Total_Students"].map("{:,.0f}".format)
district_snapshot["Total_Schools"] = district_snapshot["Total_Schools"].map("{:.0f}".format)

district_snapshot
```

## School Summary

```python
#grouped by "school_name"

grouped_complete_data = complete_data.groupby(["school_name"])

#isolated data of students passing math
#grouped by "school_name"

pass_math_data = complete_data.loc[complete_data["math_score"] >= 70]
grouped_pass_math = pass_math_data.groupby(["school_name"])

#isolated data of students passing reading
#grouped by "school_name"

pass_reading_data = complete_data.loc[complete_data["reading_score"] >= 70]
grouped_pass_reading = pass_reading_data.groupby(["school_name"])

#isolated data of students passing overall
#grouped by "school_name"

pass_overall_data = complete_data.loc[(complete_data["reading_score"] >= 70) & (complete_data["math_score"] >= 70)]
grouped_pass_overall = pass_overall_data.groupby(["school_name"])
```

```python
#used .count() to find total students
#used .first() to extract the first value in "budget"
#found student budget

per_school_type = grouped_complete_data["type"].first()
per_school_total_students = grouped_complete_data["school_name"].count()
per_school_budget = grouped_complete_data["budget"].first()
per_school_student_budget = per_school_budget / per_school_total_students

#found average math score, reading score

per_school_avg_math_score = grouped_complete_data["math_score"].mean()
per_school_avg_reading_score = grouped_complete_data["reading_score"].mean()

#found number of students passing math, reading, overall

per_school_pass_math = grouped_pass_math["school_name"].count()
per_school_pass_reading = grouped_pass_reading["school_name"].count()
per_school_pass_overall = grouped_pass_overall["school_name"].count()

#found percentage of students passing math, reading, overall

per_school_percentage_pass_math = (per_school_pass_math / per_school_total_students) * 100
per_school_percentage_pass_reading = (per_school_pass_reading / per_school_total_students) * 100
per_school_percentage_pass_overall = (per_school_pass_overall / per_school_total_students) * 100
```

```python
#created a dataframe containing school data

school_snapshot = pd.DataFrame({
    "School_Type": per_school_type,
    "Total_Students": per_school_total_students,
    "Total_School_Budget": per_school_budget,
    "Per_Student_Budget": per_school_student_budget,
    "Average_Math_Score": per_school_avg_math_score,
    "Average_Reading_Score": per_school_avg_reading_score,
    "%_Passing_Math": per_school_percentage_pass_math,
    "%_Passing_Reading": per_school_percentage_pass_reading,
    "%_Passing_Overall": per_school_percentage_pass_overall
})

school_snapshot.index.name = None

school_snapshot["Total_School_Budget"] = school_snapshot["Total_School_Budget"].map("${:,.2f}".format)
school_snapshot["Per_Student_Budget"] = school_snapshot["Per_Student_Budget"].map("${:,.2f}".format)


school_snapshot.head()
```

## Highest-Performing Schools

```python
#sorted by %_Passing_Overall in descending order

top_schools = school_snapshot.sort_values("%_Passing_Overall", ascending = False).head(5)
top_schools
```

## Lowest-Performing Schools

```python
#sorted by %_Passing_Overall in ascending order

bottom_schools = school_snapshot.sort_values("%_Passing_Overall", ascending = True).head(5)
bottom_schools
```

## Math Scores by Grade

```python
#Seperated data by grade level
#Computed the mean math_score of each
#Created a df of the data

ninth_grade = complete_data.loc[complete_data["grade"] == "9th"]
tenth_grade = complete_data.loc[complete_data["grade"] == "10th"]
eleventh_grade = complete_data.loc[complete_data["grade"] == "11th"]
twelfth_grade = complete_data.loc[complete_data["grade"] == "12th"]

ninth_grade_math = ninth_grade.groupby("school_name")["math_score"].mean()
tenth_grade_math = tenth_grade.groupby("school_name")["math_score"].mean()
eleventh_grade_math = eleventh_grade.groupby("school_name")["math_score"].mean()
twelfth_grade_math = twelfth_grade.groupby("school_name")["math_score"].mean()

math_by_grade = pd.DataFrame({
    "9th": ninth_grade_math,
    "10th": tenth_grade_math,
    "11th": eleventh_grade_math,
    "12th": twelfth_grade_math
})

math_by_grade.index.name = None

math_by_grade
```

## Reading Scores by Grade

```python
ninth_grade_reading = ninth_grade.groupby("school_name")["reading_score"].mean()
tenth_grade_reading = tenth_grade.groupby("school_name")["reading_score"].mean()
eleventh_grade_reading = eleventh_grade.groupby("school_name")["reading_score"].mean()
twelfth_grade_reading = twelfth_grade.groupby("school_name")["reading_score"].mean()

reading_by_grade = pd.DataFrame({
    "9th": ninth_grade_reading,
    "10th": tenth_grade_reading,
    "11th": eleventh_grade_reading,
    "12th": twelfth_grade_reading
})

reading_by_grade.index.name = None

reading_by_grade
```

## Scores by School Spending

```python
#created bins of school spending
#removed $ from Per_Student_Budget column
#bin based on Per_Student_Budget
#add $ to Per_Student_Budget column

spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending = school_snapshot.copy()
school_spending["Per_Student_Budget"] = school_spending["Per_Student_Budget"].apply(lambda x: x.replace("$", "")).astype(float)
school_spending["Spending_Ranges_(Per_Student)"] = pd.cut(school_spending["Per_Student_Budget"], spending_bins, labels = spending_labels)
school_spending["Per_Student_Budget"] = school_spending["Per_Student_Budget"].map("${:,.2f}".format)
```

```python
#Computed the means of relevant data
#Created df of that data

spending_math_scores = school_spending.groupby(["Spending_Ranges_(Per_Student)"])["Average_Math_Score"].mean()
spending_reading_scores = school_spending.groupby(["Spending_Ranges_(Per_Student)"])["Average_Reading_Score"].mean()
spending_passing_math = school_spending.groupby(["Spending_Ranges_(Per_Student)"])["%_Passing_Math"].mean()
spending_passing_reading = school_spending.groupby(["Spending_Ranges_(Per_Student)"])["%_Passing_Reading"].mean()
spending_passing_overall = school_spending.groupby(["Spending_Ranges_(Per_Student)"])["%_Passing_Overall"].mean()

spending_summary = pd.DataFrame({
    "Average_Math_Score": spending_math_scores,
    "Average_Reading_Score": spending_reading_scores,
    "%_Passing_Math": spending_passing_math,
    "%_Passing_Reading": spending_passing_reading,
    "%_Passing_Overall": spending_passing_overall
})

spending_summary.head()
```

## Scores by School Size

```python
#binned the data based on size

size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size = school_snapshot.copy()
school_size["School_Size"] = pd.cut(school_size["Total_Students"], size_bins, labels = size_labels)

school_size.head()
```

```python
#Computed means
#Created df

size_math_scores = school_size.groupby(["School_Size"])["Average_Math_Score"].mean()
size_reading_scores = school_size.groupby(["School_Size"])["Average_Reading_Score"].mean()
size_passing_math = school_size.groupby(["School_Size"])["%_Passing_Math"].mean()
size_passing_reading = school_size.groupby(["School_Size"])["%_Passing_Reading"].mean()
size_passing_overall = school_size.groupby(["School_Size"])["%_Passing_Overall"].mean()

size_summary = pd.DataFrame({
    "Average_Math_Score": size_math_scores,
    "Average_Reading_Score": size_reading_scores,
    "%_Passing_Math": size_passing_math,
    "%_Passing_Reading": size_passing_reading,
    "%_Passing_Overall": size_passing_overall
})

size_summary.head()
```

## Scores by School Type

```python
#Computed means
#Created df

average_math_score_by_type = school_size.groupby(["School_Type"])["Average_Math_Score"].mean()
average_reading_score_by_type = school_size.groupby(["School_Type"])["Average_Reading_Score"].mean()
average_percent_passing_math_by_type = school_size.groupby(["School_Type"])["%_Passing_Math"].mean()
average_percent_passing_reading_by_type = school_size.groupby(["School_Type"])["%_Passing_Reading"].mean()
average_percent_overall_passing_by_type = school_size.groupby(["School_Type"])["%_Passing_Overall"].mean()

type_summary = pd.DataFrame({
    "Average_Math_Score": average_math_score_by_type,
    "Average_Reading_Score": average_reading_score_by_type,
    "%_Passing_Math": average_percent_passing_math_by_type,
    "%_Passing_Reading": average_percent_passing_reading_by_type,
    "%_Passing_Overall": average_percent_overall_passing_by_type
})

type_summary.head()
```
